No presente trabalho foi utilizada a base de dados Diabetic Retinopathy Debrecen Data Set do Center for Machine Learning and Intelligent Systems. Essa base contém recursos extraídos do conjunto de imagens de Messidor para prever se uma imagem contém sinais de retinopatia diabética ou não. A extração de características das imagens são dadas da seguinte maneira:

**Atributos do banco de dados**

0) O resultado binário da avaliação da qualidade. 0 = má qualidade 1 = qualidade suficiente. 1) O resultado binário da pré-triagem, onde 1 indica anormalidade retiniana severa e 0 sua falta. 2-7) Os resultados da detecção de MA. Cada valor de característica representa o número de MAs encontrados nos níveis de confiança alfa = 0,5,. . . , 1, respectivamente. 8-15) contêm as mesmas informações que 2-7) para exsudatos. No entanto, como os exsudatos são representados por um conjunto de pontos em vez do número de pixels que constroem as lesões, esses recursos são normalizados dividindo o número de lesões com o diâmetro da ROI para compensar os diferentes tamanhos de imagem . 16) A distância euclidiana do centro de a mácula e o centro do disco óptico para fornecer informações importantes sobre a condição do paciente. Esse recurso também é normalizado com o diâmetro da ROI. 17) O diâmetro do disco óptico. 18) O resultado binário da classificação baseada em AM / FM. 19) Etiqueta da classe. 1 = contém sinais de DR (etiqueta acumulativa para as classes 1, 2, 3 de Messidor), 0 = nenhum sinal de DR.

Fonte: Retinopathy Debrecen Data Set [link text](https://archive.ics.uci.edu/ml/datasets/Diabetic+Retinopathy+Debrecen+Data+Set#)
